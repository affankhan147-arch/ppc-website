```
import os, json, csv, pathlib
base = pathlib.Path('/content/ppc-website/colab-output')
base.mkdir(parents=True, exist_ok=True)
for folder in ['raw_keyword_exports','generated_briefs','cleaned_csvs','summaries_for_codex']:
    (base/folder).mkdir(parents=True, exist_ok=True)
print('Colab output folders ready:', base)
```

```
keywords = [
    'emergency plumber Dallas',
    '24 hour plumber Dallas',
    'burst pipe Dallas',
    'sewer backup Dallas',
    'water heater leaking Dallas',
    'drain cleaning Dallas',
    'plumbing emergency cost',
    'emergency plumber cost',
]
with open(base/'cleaned_csvs'/'keyword_seed_list.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['keyword','intent','priority'])
    for k in keywords:
        writer.writerow([k,'call-intent','high'])
print('Created keyword_seed_list.csv')
```

```
summary = '# Pre-Codex Colab Summary\n\nColab prepared keyword seed data and output folders.\nCodex should read ops/codex_minimum_context.md and command-center final command first.\nNo paid ads. Organic calls first. No fake GMBs, fake reviews, fake addresses, or copied competitor images.\n'
out = pathlib.Path('/content/ppc-website/ops')
out.mkdir(exist_ok=True)
(out/'pre_codex_colab_summary.md').write_text(summary, encoding='utf-8')
print('Created ops/pre_codex_colab_summary.md')
```
