```python
from pathlib import Path
import csv
import json
from datetime import datetime, timezone

BASE = Path('/content/ppc-workflow-support')
OUT = BASE / 'colab-output'
folders = [
    OUT / 'cleaned_csvs',
    OUT / 'generated_briefs',
    OUT / 'content_outlines',
    OUT / 'summaries_for_codex',
    OUT / 'project_support_data',
]
for folder in folders:
    folder.mkdir(parents=True, exist_ok=True)

print('Support output ready:', OUT)
```

```python
keyword_rows = [
    {'keyword': 'emergency plumber near me', 'intent': 'urgent-call', 'priority': 'high', 'notes': 'Use only where service area is accurate.'},
    {'keyword': '24 hour plumber', 'intent': 'urgent-call', 'priority': 'high', 'notes': 'Avoid fake local office claims.'},
    {'keyword': 'burst pipe repair', 'intent': 'service-problem', 'priority': 'high', 'notes': 'Explain response expectations honestly.'},
    {'keyword': 'water heater leaking', 'intent': 'service-problem', 'priority': 'medium', 'notes': 'Include safety warning and call CTA.'},
    {'keyword': 'sewer backup help', 'intent': 'service-problem', 'priority': 'medium', 'notes': 'Include cleanup and plumber distinction.'},
    {'keyword': 'drain cleaning service', 'intent': 'service-page', 'priority': 'medium', 'notes': 'No unsupported price claims.'},
    {'keyword': 'plumbing emergency cost', 'intent': 'informational', 'priority': 'medium', 'notes': 'Use ranges only when verified later.'},
]

keyword_path = OUT / 'cleaned_csvs' / 'keyword_seed_list.csv'
with keyword_path.open('w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=['keyword', 'intent', 'priority', 'notes'])
    writer.writeheader()
    writer.writerows(keyword_rows)

print('Created', keyword_path)
```

```python
brief = '''# Emergency Plumbing Page Brief

## Goal
Create a helpful organic page that answers urgent plumbing questions and encourages a call without fake local claims.

## Search Intent
The visitor needs fast help, cost context, and confidence that the next step is a phone call.

## Required Sections
- Immediate problem summary
- Common emergency examples
- What to do before calling
- Honest service-area language
- FAQ / AEO answers
- Clear call CTA

## Compliance Notes
Do not claim physical offices, licenses, insurance, reviews, or response times unless the owner verifies them.
'''

brief_path = OUT / 'generated_briefs' / 'emergency_plumbing_page_brief.md'
brief_path.write_text(brief, encoding='utf-8')
print('Created', brief_path)
```

```python
outline = '''# Service Page Content Outline

1. H1: Emergency plumbing help
2. Short answer: what the page helps with
3. Call CTA
4. Common emergency situations
5. What information to have ready before calling
6. Service-area note without fake office claims
7. FAQ answers for AEO
8. Final call CTA

Safety reminder: do not invent reviews, offices, licenses, insurance, or guaranteed arrival times.
'''

outline_path = OUT / 'content_outlines' / 'service_page_outline.md'
outline_path.write_text(outline, encoding='utf-8')
print('Created', outline_path)
```

```python
support_data = {
    'project': 'PPC lead generation platform',
    'phase': 'support preparation only',
    'generated_at_utc': datetime.now(timezone.utc).isoformat(),
    'safe_outputs': ['keyword CSV', 'page brief', 'content outline', 'summary'],
    'blocked_outputs': ['ads', 'fake profiles', 'fake reviews', 'credentials', 'payment data'],
}

json_path = OUT / 'project_support_data' / 'project_support_data.json'
json_path.write_text(json.dumps(support_data, indent=2), encoding='utf-8')
print('Created', json_path)
```

```python
summary = '''# Pre-Codex Colab Summary

Colab prepared support-only files for the PPC workflow.

Generated files:
- cleaned_csvs/keyword_seed_list.csv
- generated_briefs/emergency_plumbing_page_brief.md
- content_outlines/service_page_outline.md
- project_support_data/project_support_data.json

Codex should treat these as support inputs only and must still follow ops/api_safety_policy.md.
'''

summary_path = OUT / 'summaries_for_codex' / 'pre_codex_colab_summary.md'
summary_path.write_text(summary, encoding='utf-8')
print('Created', summary_path)
```
